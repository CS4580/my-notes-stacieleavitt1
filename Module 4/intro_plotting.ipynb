### Bar Plot
Useful to show or compare segments of information or categorical data

```python
# Load cities.csv data
cities_df = pd.read_csv('../data/cities.csv')
cities_df.head()
```

```python
# Filter records from Year = 2019
cities_2019df = cities_df[cities_df['Year']==2019]
cities_2019df.head()
```

```python
# Now create a bar chart of the population in each city in 2019
cities_2019df.plot(kind='bar', x='City', y='Population', title='Population of Cities in 2019', figsize=(10,6))
```

```python
# You may want to plot data horizontally
# Since we are only plotting one set of values we don't need a legend
cities_2019df.plot(kind='barh', x='City', y='Population', title='Population by Cities in 2019', legend=False, figsize=(10,6))
```

### Scatter Plots
When you are interested in the relationship between two variables, scatter plots are probably the best choice

```python
# Load data for houses.csv
houses_df = pd.read_csv('../data/homes.csv')
houses_df.head()
```

```python
# Plot the scatter plot
houses_df.plot(kind='scatter', x='List', y='Rooms')
```

```python
# Filter data directly
houses_df[houses_df['Age'] < 50].plot(kind='scatter', x='Age', y='List')
```

```python
# If you want to eliminate outliers, you can filter the data before plotting
houses_plot_df = houses_df[houses_df['Age'] < 50]
houses_plot_df = houses_plot_df[houses_plot_df['List'] < 350]
# Now plot
houses_plot_df.plot(kind='scatter', x='Age', y='List')
```

### Histograms
Use to represent frequency in noncategorical data or data that is numeric in value. Heights, price, age, etc


```python
houses_plot_df.head()
```

```python
# If you want to know the distribution of acreage for the houses
# sold in the filtered dataset, use a histogram
houses_plot_df['Acres'].plot(kind='hist', title='Acre Distribution for Homes age < 50 and list < 350k')
```

### Area Plot
Useful when you want to understand how values develop and are composed over time

```python
## Load data set
enrollments_df = pd.read_csv('../data/enrollment.csv')
enrollments_df.head()
```

```python
# Create an area plot of the enrollments
enrollments_df.plot(kind='area', x='Year')
```

```python
# Data is not good because it is not organized by year
# We want to see the total, and how it changes over time
# Sort the data by year, then plot it using the pivot method in pandas
pivot_df = enrollments_df.pivot(index='Year', columns='Class', values='Enrollment')
pivot_df.head()
```

```python
# Plot it
pivot_df.plot(kind='area')
```

```python
# The data is stacked by default.
# You can set this option to false, stacked=False
pivot_df.plot(kind='area', stacked=False)
```
