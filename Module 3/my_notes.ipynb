## Identify missing values

````python
# Find the number of missing values
df_data.isnull().sum()
````

````python
# Missing values by percentage
(df_data.isnull().sum()/len(df_data))*100
````

````python
# Remove the company column or feature
df_data.drop('company', inplace=True, axis=1)
````

````python
df_data.head()
````

````python
# Fill children with zeros
df_data.children = df_data.children.fillna(0)
````

````python
# Missing values by percentage
(df_data.isnull().sum()/len(df_data))*100
````

````python
# Unique values in the country feature
df_data['country'].value_counts()
````

````python
# Drop all rows with missing data values in the country feature
df_data = df_data.dropna(subset=['country'])
````

## Duplicates

Practice how to handle duplicate values


````python
# Total number of duplicates in the dataset. (identical data every single column)
df_data.duplicated().sum()
````

````python
# Extract the dataframe containing the duplicate values
df_data.loc[df_data.duplicated(), :]
````

````python
# drop the duplicate records, tell it which to keep
df_data.drop_duplicates(inplace=True, keep='first')

# Print the shape of the dataframe to see if it changed
df_data.shape
````

## Different Datatypes

### Datetime
Convert the column "object" to a given type, in this case, datetime

Get the year, month, and day. Make 3 columns from datetime colum

Now, we can drop the temporary column datetime

````python
df_data['reservation_status_date'] = pd.to_datetime(df_data['reservation_status_date'])
df_data.info()
````

````python
df_data['reservation_status_year'] = df_data['reservation_status_date'].dt.year
df_data['reservation_status_month'] = df_data['reservation_status_date'].dt.month
df_data['reservation_status_day'] = df_data['reservation_status_date'].dt.day
df_data.head()
````

````python
df_data.drop('reservation_status_date', inplace=True, axis=1)
````
