One of the most common magic commands is the `%matplotlib inline` command. This command will enable plots to be displayed directly inside the notebook.

````python
%matplotlib inline
````

````python
"""Simple demo of scatter plot"""
import numpy as np
import matplotlib.pyplot as plt

n = 50 
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
# Calculate the area
area = np.pi * (15*np.random.rand(n))**2  # 0 to 15 pt radius
# Plot data
plt.scatter(x,y, s = area, c = colors, alpha = 0.5)
plt.show()
````

# Conda

Create and manage your virtual environments.

To create virtual env. with a specfic `python` version. 

```bash
conda create -n cs4580 python=3.10
```
Do this in Powershell or command prompt (might work better), it will ask you to bring in new packages, say yes.
Then To activate your environment, first run (sometimes):
```bash
conda init
```
then

```bash
activate cs4580 #or 
conda activate cs4580
```
Using ```conda env list``` will show your different environments
Code ```conda list``` Gets you the list of current packages installed


# Numpy

## Features
- High Performance
- Vectorization
- Broadcasting
- Interoperability

Less loops so more speed, especially thanks to vectorization

### Comparison with other alternatives

#### Python list vs Numpy Arrays
- Elements of different types `vs` same type
- Variable size nesting is possible `vs` same size
- No operations at the list level `vs` operations allowed at the array level
- Uses a lot of memory `vs` less memory
