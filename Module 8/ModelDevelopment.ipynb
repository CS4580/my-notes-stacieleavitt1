## Linear Regression
A method to help us understand the relationship between two variables:
- The predictor/independent variable (x)
- The response/dependent variable (y)

The result of a linear regression is a function that predicts the response variable as a function of the predictor

$
\widehat{Y} = a + bX
$

- `a` refers to the `intercept` of the line
- `b` refers to the `slope`

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create a linear regression object
lm = LinearRegression()
lm
```

## How could the `sqft_living` help us predict the house `price`?

```python
X = df[['sqft_living']] # This needs to be a 2D array
Y = df['price'] # This can be 1D
```

```python
# Fit linear model
lm.fit(X, Y)
```

```python
# Get the output prediction using X
Yhat = lm.predict(X)
# Print first 5 results
Yhat[:5]
```

```python
# Get the value of the intercept (a)
lm.intercept_
```

```python
# Get the value of the slope (b)
lm.coef_
```

Final Estimated linear model

$
\widehat{Y} = a + bX
$

In our case:
$
price = -43580.74 + 280.62356X
$

## Task 1
Repeat the process but using `bedrooms` as the independent variable and `price` as your dependent

```python
# Create a linear regression object
lm2 = LinearRegression()

X = df[['bedrooms']] # This needs to be a 2D array
Y = df['price'] # This can be 1D

# Fit linear model
lm2.fit(X, Y)

# Get the value of the intercept (a)
lm2_a = lm2.intercept_
print(f'The intercept: {lm2_a}')

# Get the value of the slope (b)
lm2_coef = lm2.coef_
print(f'The slope is: {lm2_coef}')

Yhat2 = lm2_a + lm2_coef*X

print(f'The linear model equation: price = {lm2_a} + {lm2_coef}bedrooms')
print(f'The predictions of yhat:\n {Yhat2}')
```

## Multiple Linear Regression
If you want to use more variables in our model to predict the price, then, use a `multiple linear regression`

$
\widehat{Y} = a + b_{1}X_{1} + b_{2}X_{2} + b_{3}X_{3}...b_{n}X_{n}
$

For this example, good predictors of price will be:
- sqft_living
- bedrooms
- bathrooms
- sqft_lot

```python
# Reload Data & imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
# Load data
df = pd.read_csv("../data/kc_house_data.csv")
```

```python
# Get all the predictor variables
X = df[['sqft_living', 'bedrooms', 'bathrooms', 'sqft_lot']] # independent variables will need to be 2D
Y = df['price'] # Dependent variable only 1D
```

```python
lm3 = LinearRegression()
lm3.fit(X,Y)
```

```python
a = lm3.intercept_
bs = list(lm3.coef_)
print(f'a: {a} and slopes: {bs}')

Yhat3 = a + (bs[0]*X['sqft_living']) + (bs[1]*X['bedrooms']) + (bs[2]*X['bathrooms']) + (bs[3]*X['sqft_lot'])
print(Yhat3)
```

## Model Evaluation Using Visualization
### Regression Plot

Regression plots are a gerat tool to plot Linear Regression. This plot shows a combination of a scatterplot and a linear regression line

```python
width = 10
height = 6
# plot the figure
plt.figure(figsize=(width, height))
sns.regplot(x='sqft_living', y='price', data=df)
plt.ylim(0,) # ylim is used to set the limit of the y-axis
```

```python
# Compare now versus the price
plt.figure(figsize=(width, height))
sns.regplot(x='yr_built', y='price', data=df)
plt.ylim(0,)
```

### Residual Plot
This is a good way to visualize the `variance` of the data.
- `Residual` is the difference between the `observed value (y)` and the `predicted value (yhat)`


```python
plt.figure(figsize=(width, height))
sns.residplot(x = df['sqft_living'], y = df['price'])
```
